# Find Null Values & Fill it (if require)

1) Drop Raw which has NaN Values for all attributes
2) Drop Raw of Comodity attributes which has NaN Value as we cannot get any accurate insights whithout comodity details  
3) Fill NaN Retail price value with Interpolate Values

# Load Clean DataSet

```python
df_Clean = pd.read_csv("Monthly_Retails_Price_Clean.csv")
df_Clean
```

```python
#Change datatype of Date Fetures
df_Clean['Date']= df_Clean['Date'].astype('datetime64[ns]')
```

```python
df_Clean.info()
```

```python
df_Clean.dtypes
```

```python
#Add Year Features 
df_Clean['Year'] = df_Clean['Date'].dt.year
df_Clean
```

```python
df_Clean = df_Clean.drop("Unnamed: 0",axis=1)
df_Clean
```

```python
df_Clean.isnull().sum()
```

```python
df_Clean.Commodity.unique()
```

# Which Date has highest price of wheat?

```python
df2_list = ['Wheat']
Commodity_Wheat= df_Clean[df_Clean.Commodity.isin(df2_list)]
Commodity_Wheat
```

```python
Commodity_Wheat.max()
```

```python
Datewise_Wheat_Price= Commodity_Wheat.groupby(['Commodity','Year'])['Retail Price'].max()
Datewise_Wheat_Price
```

# Which is the highest price of wheat state wise?

```python
Statewise_Wheat_Price= Commodity_Wheat.groupby(['Commodity','State'])['Retail Price'].max()
Statewise_Wheat_Price
```

```python
Statewise_Wheat_Price.plot(kind="bar",figsize=(16,4))
plt.title("Highest Price of Wheat Statewise")
plt.show()
```

# What is the Pomphret price in kerala from 2001 to 2021?
