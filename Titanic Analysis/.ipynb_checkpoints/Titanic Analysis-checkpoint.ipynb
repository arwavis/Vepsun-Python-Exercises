### Count Survived & Non Survived and show on pie chart

```python
survived = data_frame["Survived"].value_counts()
survived
```

```python
plt.pie(survived, labels=["Not Survived", "Survived"], autopct= "%0.2f %%", colors = ["red", "green"], explode= (0,0.1),shadow =True)
plt.title ("Survived and Not Survived Passengers")
plt.legend(loc = "lower left")
plt.show
```

### Find out how many female passengers had travelled in first class and show on pie  chart

```python
female_class=data_frame[(data_frame.Sex=="female")].Pclass.value_counts()
female_class
```

```python
plt.pie(female_class, labels=["3rd", "1st", "2nd"], autopct= "%0.2f %%",explode=(0,0.1,0.1),shadow =True)
plt.title ("Female Passengers")
plt.legend(loc = "lower left")
plt.show
```
