### Activity Data Vs Calories Burnt

```python
### Activity Data Vs Calori Burntplt.figure(figsize=(12,6))
sns.barplot(x = 'ActivityDate', y = 'Calories', data=df)
# Allows you to rotate the X values to 90 degree.
plt.xticks(rotation=90)
plt.show()
```

### Activity Data Vs Total Steps

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'ActivityDate', y = 'TotalSteps', data=df)
# Allows you to rotate the X values to 90 degree.
plt.xticks(rotation=90)
plt.show()
```

```python
Distance = df[['TotalSteps','TotalDistance']]
Distance
```

```python
Distance.plot(kind="bar",figsize=(16,4))
plt.title("Total Distance comparing Total Steps")
plt.show()
```
